```python
def plot_data(df, combined_list, p):
    for element in combined_list:
        if element in df.columns:
            # Convert to numeric and drop NaN values
            df[element] = pd.to_numeric(df[element], errors='coerce')
            df[element].dropna(inplace=True)  # Or handle NaN as needed
            
            if not df[element].empty:  # Check if the series is not empty
                plt.figure(figsize=(20, 10))
                df[element].plot(kind='line', marker='o')
                plt.title(f"Data for {element} in {p}")
                plt.xlabel("Index")  # Modify as needed
                plt.ylabel(f"{element} Values")  # Modify as needed
                plt.grid(True)
                
                # Save the plot as an image file
                plt.savefig(f"{p}/{element}_graph.png")
                plt.close()
            else:
                print(f"No numeric data available for {element} in DataFrame")
        else:
            print(f"Column {element} not found in DataFrame")
```
